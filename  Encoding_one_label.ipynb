# One Hot Encoding

one hot encoding we have data limited like yes,no and marrried,unmeride,and urban, semiarban,rular etc,

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
 dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
```

```python
one hot encoding we can perform with the help of get dummies pandas library
and sklear one hot encoder
```

```python
# pandas library perform one hot encoding
```

```python
en_data = dataset[["Gender","Married"]]
```

```python
pd.get_dummies(en_data).info()
```

```python
# sklearn library one hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first")
arr = ohe.fit_transform(en_data).toarray() # it give data in the form of sparse matrix 
# in sparse matrix data is almost fill with one and zero
```

```python
pd.DataFrame(arr,columns=["Gender_Male","Married_Yes"])
```
