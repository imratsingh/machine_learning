# Standardization

-- Standardization is a feature scaling technique that transforms data so that it has a mean of 0 and a standard deviation of 1.

Outliers more than

Xnew = (Xi - Xmean) / std

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicatIncome_ss"] =pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"]) # it give data in the form of array
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Befor")
sns.distplot(dataset["ApplicatIncome_ss"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

# Normalization


Normalization is a feature scaling technique that transforms data into a fixed
range, usually between 0 and 1

Normalization → Range = 0 to 1

KNN / Neural Network

Xnew = (Xi - min(X)) /(max(X) - min(X))

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
```

```python
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("befor")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["CoapplicantIncome_min"])
plt.show()
```
